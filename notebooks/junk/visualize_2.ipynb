```python
#p_complete['stationary_prob_normalized'] = 1/(p_complete[p_complete.topic_name=='living conditions']['stationary_prob'].values[0] / p_complete['stationary_prob']) 
wfreqs = p_women[0:75].set_index('topic_name')['stationary_prob'].to_dict()
```

```python
mfreqs = p_men[0:75].set_index('topic_name')['stationary_prob'].to_dict()
```

```python
def black(word, font_size, position, orientation,
    font_path, random_state):
    if (word =="social bonds") or (word =="aid giving") or (word =="friends") or (word =="food sharing"):
        return 'red'
    elif (word =="tattoo") or (word =="clothing") or (word =="Appell") or (word == "escapes"):
        return 'blue'`
    else:
        return 'black'
```

```python
wordcloud_w = WordCloud(width = 800, height = 800, 
                background_color ='white', color_func=black,
                min_font_size = 8,max_font_size=40).generate_from_frequencies(wfreqs)
```

```python
wordcloud_m = WordCloud(width = 800, height = 800, 
                background_color ='white', color_func=black,
                min_font_size = 8,max_font_size=40).generate_from_frequencies(mfreqs)
```
