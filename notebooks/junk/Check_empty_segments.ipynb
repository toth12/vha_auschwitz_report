```python
new_input_data_set=[]
removed = 0
added = 0
# Eliminate empty steps
for f,interview in enumerate(input_data_set):
    print ('\n')
    print('Int number'+str(f))
    print ('\n')
 
    new_interview = []
    non_consecutive = 0
    for c,row in enumerate(interview):
        if row.sum()>0:
            new_interview.append(row)
            added +=1
            non_consecutive = 0
        else:
            removed +=1
            non_consecutive +=1
            print ('step'+str(c))
    new_input_data_set.append(np.vstack(new_interview))
    
```
